# Plasma parameters

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import pi, e, hbar, m_e, c, epsilon_0
```

```python
# electron beam parameters
E0 = 0.511 # MeV
E = 250 # MeV
epsn = 5e-6 # initial emittamce

gamma = E/E0+1

density = np.arange(1, 1001, 1)
density = density*1e14*1e6
```

```python
wp = np.sqrt(density * e**2 / epsilon_0 / m_e)
lamda = 2 * pi * c/ wp
kp = wp/c
skin_depth = 1 / kp
skin_depth *= 1e6


E_WB = c*m_e*wp/e # wavebreaking field

plasma_matched_radius = ( 2*epsn**2/gamma/kp**2 ) ** (1./4.)
```

```python
figsize = (5, 3)  # Taller height for a 3-row vertical layout
fig, ax = plt.subplots(figsize=figsize)  # 3 rows, 1 column

# Repeat the process for each subplot:
###############################################
# Plot 1: Witness Energy
ax.tick_params(which='both', width=1.2)
ax.tick_params(which='major', length=5)
ax.tick_params(which='minor', length=2)

ax.plot(density/1e6, skin_depth, color='r', linestyle='-', linewidth=2, label='skin depth')


ax.set_xscale('log')
ax.set_xlim(1e14, 1e17)  # Ensure xlim is correctly applied
ax.set_ylim(10, 600)  # Ensure xlim is correctly applied
#ax.set_xticks([6e15, 1e16, 2e16, 3e16])
#ax.get_xaxis().set_major_formatter(plt.ScalarFormatter())
#ax.ticklabel_format(style='sci', axis='x', scilimits=(0, 0))
ax.set_yscale('log')
#ax.set_ylim(250, 650)
#ax.axhline(y=500, color='k', linestyle='--', linewidth=2)
ax.set_ylabel(r'$ k_\mathrm{p}^{-1}\,  \mathrm{[\mu \, m]}$', fontsize=12)
ax.set_xlabel(r'$ n_\mathrm{p} \,  \mathrm{[cm^{-3}]}$', fontsize=12)
#ax.text(0.1, 0.9, '(a)', transform=ax.transAxes, fontsize=fontsize14, verticalalignment='top')
#ax.legend(ncol=2, loc='lower right', fontsize=12)
#ax.yaxis.set_major_locator(MultipleLocator(100))  # Minor ticks every 5 units
#ax.yaxis.set_minor_locator(MultipleLocator(50))  # Minor ticks every 5 units
ax.tick_params(axis='both', which='major', labelsize=14)  # Change 14 to your desired font size
plt.savefig('plasma1.pdf', dpi=100, bbox_inches='tight')
```
